```python
results = pd.read_csv('1976-2018-house.csv', encoding="ISO-8859-1")
```

```python
df = results[results['year'] > 2009]
```

```python
len(df)
```

```python
df.candidatevotes = df.candidatevotes.str.replace(",","")
```

```python
df.candidatevotes = df.candidatevotes.astype(float)
```

```python
df['district'] = df['district'].apply(lambda x: '{0:0>2}'.format(x))
```

```python
df.head()
```

```python
df['DISTRICT'] = df.state_po + "-" + df.district
```

```python
df.head()
```

```python
byrace = df.groupby(['DISTRICT', 'year', 'party'], sort=False)['candidatevotes'].max()
```

```python
byrace.tail()
```

```python
byrace.to_csv('byrace.csv')
```

```python
byrace2 = pd.read_csv('byrace.csv')
```

```python
byrace3 = byrace2.loc[byrace2.reset_index().groupby(['DISTRICT', 'YEAR'])['VOTES'].idxmax()] 
```
